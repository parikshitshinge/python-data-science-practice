# Using Palmer Archipelago penguin dataset for analysis
#### This data set includes 3 species of penguins with their characteristics such as culemen length, culmen depth, flipper length, body mass etc. 
#### The data set also tells us which island each penguin was found and its gender

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
original_data = pd.read_csv("penguins_size.csv")
original_data
```

```python
original_data.describe()
```

```python
original_data.groupby('species').describe()
```

```python
sns.pairplot(original_data, hue='species')
```

## Observations
#### 1: It is relatively easy to classify Penguins' species based on their culmen length & culmen depth
Adelie possesses lowest culmen length than Chinstrap & Gentoo.
Gentoo possesses lowest culmen depth than Adelie & Chinstrap and higher culmen length than Adelie
#### 2: Gentoo has highest flipper length than Adelie & Chinstrap
#### 3: Gentoo possesses higher body mass than Adelie & Chinstrap while body mass of Adelie & Chinstrap are same

```python
sns.pairplot(original_data, hue='sex')
```

#### From above pairplot, it is observed that MALE penguines have larger culment length, culmen depth, flipper length & body mass as well for each species

```python
sns.boxplot(x='species', y='culmen_length_mm', data=original_data)
```

#### Adelie has smallest culmen by length

```python
sns.boxplot(x='species', y='culmen_depth_mm', data=original_data)
```

#### Gentoo has smallest culmen by depth

```python
sns.scatterplot(x='culmen_length_mm', y='culmen_depth_mm', hue='species', data=original_data)
```

```python
sns.scatterplot(x='culmen_length_mm', y='culmen_depth_mm', hue='sex', data=original_data)
```

```python
sns.pairplot(original_data, hue='island')
```

#### Biscoe island has penguins population with highest flipper length & lowest culment length which is Gentoo species
#### Other 2 species - Adelie & Chinstrap are found to be on Torgersen & Dream islands
