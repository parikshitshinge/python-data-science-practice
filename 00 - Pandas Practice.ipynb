# Operations on DataFrame

```python
df_weather = pd.read_csv("weather_data_cities.csv")
df_weather
```

```python
df_weather.describe()
```

```python
df_grouped_city = df_weather.groupby('city')
```

```python
df_grouped_city.describe()
```

```python
mumbai_weather = df_weather[df_weather['city']=='mumbai']
ny_weather = df_weather[df_weather['city']=='new york']
paris_weather = df_weather[df_weather['city']=='paris']
```

```python
mumbai_weather
```

```python
ny_weather
```

```python
paris_weather
```

```python
# extend data frame with rows - row union
mumbai_ny_weather_row = pd.concat([mumbai_weather, ny_weather])
mumbai_ny_weather_row
```

```python
# extend data frame with columns - column union
mumbai_ny_weather_col = pd.concat([mumbai_weather, ny_weather], axis=1)
mumbai_ny_weather_col
```

# Accessing data using index

```python
df_weather.loc[2] # data at index 2
```

```python
df_weather.loc[:10] # data from index 0 to 10
```
