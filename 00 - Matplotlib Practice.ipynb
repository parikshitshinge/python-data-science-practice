```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4,5])
plt.xlabel("Numbers")
plt.ylabel("Indices")
plt.show()
```

```python
# show grid on plot
plt.grid()
plt.show()
```

# Show data points with different shapes & colors

```python
plt.grid()
plt.plot([1,2,3], [1,4,9], 'ro') # r=red, o= o shape
plt.show()
```

```python
plt.grid()
plt.plot([1,2,3], [1,4,9], 'b--') # b=blue, --= -- dashed line
plt.show()
```

```python
plt.grid()
plt.plot([1,2,3], [1,4,9], 'rs') # r=red, s= square shape
plt.show()
```

```python
plt.grid()
plt.plot([1,2,3], [1,4,9], 'g^') # g=green, ^= triangle shape
plt.show()
```

```python
lst = [1,2,3,4,5]
```

```python
plt.plot(lst, [ele**2 for ele in lst], 'rs', label='^2')
plt.plot(lst, [ele**3 for ele in lst], 'bx', label='^3')
plt.plot(lst, [ele*2 for ele in lst], 'g^', label='x2')
plt.plot(lst, [ele*3 for ele in lst], 'yo', label='x3')
plt.show()
```

# Formatting lines

```python
x1 = [2,4,6]
y1 = [1,2,3]
x2 = [4,8,12]
y2 = [4,5,6]

lines = plt.plot(x1,y1,x2,y2)
# format individual lines

plt.setp(lines[0], color='r', linewidth=2)
plt.setp(lines[1], color='b', linewidth=5, linestyle='--')

plt.show()
```

# Plotting subplots

```python
plt.subplot(211) # 2: no of rows, 1: no of columns, 1: below code is for 1st figure
plt.grid()
plt.plot(x1,y1,'b--')

plt.subplot(212) # 2: no of rows, 1: no of columns, 2: below code is for 2nd figure
plt.plot(x2,y2,'r-')

plt.show()
```

### Subplot 4 graphs

```python
plt.subplot(221)
plt.plot(x1,y1,'b--')
plt.grid()

plt.subplot(222)
plt.plot(x1,y2,'b-')

plt.subplot(223)
plt.plot(x2,y1,'r--')

plt.subplot(224)
plt.plot(x2,y2,'r-')
plt.grid()

plt.show()
```
